```python
import re
from textblob import TextBlob
import pandas as pd
```

```python
data = pd.read_csv('dataset/clean_binus.csv')
```

```python
data.sample(10)
```

```python
# Fungsi analisis sentimen
def get_tweet_sentiment(tweet):
    analysis = TextBlob(tweet)
        
    try:
        analysis = analysis.translate(from_lang='id', to='en')
        print('Translated')
    except:
        analysis = analysis
    
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'
```

```python
sentiment = []
for tweet in data.clean_tweet:
    sentiment.append(get_tweet_sentiment(tweet))
    
# Membuat kolom baru clean_tweet
col_sentiment = pd.DataFrame({'sentiment':sentiment})
data = pd.concat([data,col_sentiment],axis=1)
```

```python
data.sample(10)
```

```python
positiveTweets = data[data['sentiment'] == 'positive']
print("Positive tweets percentage: {} %".format(100*len(positiveTweets)/len(data)))
```

```python
negativeTweets = data[data['sentiment'] == 'negative']
print("Positive tweets percentage: {} %".format(100*len(negativeTweets)/len(data)))
```
